## Pandas Implementation

```python
import pandas as pd

## Create a DataFrame from a dictionary
data = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, 30, 35],
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}
#Your code here
data=pd.DataFrame(data)

#Display the first 2 rows of the data frame
#Your code here
data.head(2)

#Print the age column
#Your code here
print(data['Age'])

#Filter rows where age is greater than 26
#Your code here
data[data['Age']>26]

#Add a new column 'Country' with the value 'India' for all rows
#Your code here
data['Country']='India'

data1 = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, None, 35],
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}

df2= pd.DataFrame(data1)

# Fill missing values in the 'Age' column with the mean age
#Your code here
df2['Age'].fillna(df2['Age'].mean(),inplace=True)
```

## Matplotlib Implementation

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 3, 4])
ypoints = np.array([2, 4, 6, 8])

# Plot these points without drawing a line
plt.figure(figsize=(6, 6))
plt.plot(xpoints, ypoints, 'o')
plt.title('Plot of points without line')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()

# Plotting with marker: Plot these points with a star marker
plt.figure(figsize=(6, 6))
plt.plot(xpoints, ypoints, marker='*', linestyle='None', color='green')
plt.title('Plot of points with star marker')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()

# Using fmt format, add circular marker, red color and dashed line
plt.figure(figsize=(6, 6))
plt.plot(xpoints, ypoints, 'ro--', linewidth=2)
plt.title('Plot of points with circular marker, red color, and dashed line')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()

# Create a scatter plot for xpoints and ypoints
plt.figure(figsize=(6, 6))
plt.scatter(xpoints, ypoints)
plt.title('Scatter plot of xpoints and ypoints')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()

# Set color to the scatter plot: Blue, Green, Red, and Yellow color for each point respectively
colors = ['blue', 'green', 'red', 'yellow']
plt.figure(figsize=(6, 6))
for i in range(len(xpoints)):
    plt.scatter(xpoints[i], ypoints[i], color=colors[i], label=f'Point {i+1}')
plt.title('Scatter plot with colored points')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```
